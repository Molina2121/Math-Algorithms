**Ejercicio 3.** Considera la curva $\mathbf{y}(t)=(\cos^3 t, \operatorname{sen}^3 t, \frac{3}{4}\cos 2t)$.

1.  Encuentra el periodo $T$ de esta curva y encuentra dónde es regular.
2.  Calcula el triedro de Frenet-Serret donde sea birregular, la curvatura y la torsión.
3.  Dibuja la curva y el triedro con una animación.
4.  Calcula la longitud, curvatura total y torsión total.

La curvatura total y la torsión total se definen así:
$$
\kappa^{\textrm{tot}}=\int_0^T \!\!\!\!\!\!\kappa_\mathbf{y}(t)\operatorname{d}t,\qquad
\tau^{\textrm{tot}}=\int_0^T \!\!\!\!\!\!\tau_\mathbf{y}(t)\operatorname{d}t.
$$

$\textbf{Apartado 1}$

```python
reset ()
var('x,y,z,t,s,a,b,c,t1,t2,t3',domain='real')
V=vector([cos(t)^3,sin(t)^3,(3/4)*cos(2*t)])
```

```python
def curvaespacial(V,vr=t):
    V1,V2,V3=[V.derivative(vr,j) for j in (1,2,3)]
    V12=V1.cross_product(V2)
    reg=(V1*V1).simplify_full()
    ct=V1.normalized().canonicalize_radical().simplify_full()
    cn=ct.derivative(vr).normalized().simplify_full()
    cb=V12.normalized().canonicalize_radical().simplify_full()
    kappa=V12.norm()/V1.norm()^3
    kappa=kappa.canonicalize_radical().simplify_full()
    tau=Matrix([V1,V2,V3]).det()/(V12*V12)
    tau=tau.simplify_full()
    return V1,reg,ct,cn,cb,kappa,tau
```

```python
V1,reg,ct,cn,cb,kappa,tau=curvaespacial(V)
show(V1,reg,ct,cn,cb,kappa,tau)
```

```python
P=parametric_plot3d(V,(t,0,2*pi), thickness = '2')
P
```

Por un lado, el período de $\cos (2t) $ es $\pi$. Por otro, los períodos de $\cos^3 (t)$ y $\sin^3 (t)$ es $2 \pi$. 

$\to $ Por tanto, el período de nuestra curva será $2 \pi$.
