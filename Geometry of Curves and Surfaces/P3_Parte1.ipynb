## Ejercicio 3
El teorema fundamental de curvas planas nos dice que dada una función diferenciable $\kappa:[a,b]\to\mathbb{R}$ 
existe una única curva plana $\mathbf{x}:[a,b]\to\mathbb{R}^2$ parametrizada por el arco y tal que $\mathbf{x}(a)=(0,0)$, $\dot{\mathbf{x}}(a)=(1,0)$ y su curvatura es $\kappa$. 
Busca lo que son las integrales de Fresnel, encuentra curvas para $\kappa=t,t+1$ y relaciónalas.

*Usa `fresnel_cos(t)` y `fresnel_sin(t)` para poder usar propiedades de funciones.*


$\textbf{Busca lo que son las integrales de Fresnel.}$

Las *integrales de Fresnel* $S(x)$ y $C(x)$ son dos funciones trascendentales que se definen según las siguientes expresiones integrales
$$S(x)=\int_{0}^x \sin(t^2)dt ~~~, ~~~C(x)=\int_{0}^x \cos(t^2)dt$$

```python
reset()
x = var('x t')

S(x) = fresnel_sin(x)
C(x) = fresnel_cos(x)

S = plot(S(x), (x, 0, 5), color='red', legend_label='S(x)')
C = plot(C(x), (x, 0, 5), color='green', legend_label='C(x)')

(S + C).show(figsize = 3)
```

Notamos que las funciones empleadas, `fresnel_cos(t)` y `fresnel_sin(t)`, hacen referencia a las integrales normalizadas de Fresnel.

$\textbf{Encuentra curvas para $\kappa=t,t+1$ y relaciónalas.}$ 

Sea $\kappa_1=t$. Vamos a emplear el Teorema fundamental de curvas planas, sabemos que existe una única curva plana $\textbf{x}_1:\mathbb{R}\to\mathbb{R}$ parametrizada por el arco tal que su curvatura es $\kappa_1=t$.

Dado que $\kappa_1(s)=\frac{d\theta}{ds}$, vamos a obtener $\theta(s)$
$$\kappa_1(s)=\frac{d\theta}{ds} ~~~ \Longleftrightarrow ~~~ d\theta=\kappa_1(s)ds ~~~\Longleftrightarrow ~~~ \theta(s)=\int \kappa_1(s)ds$$

Puesto que $\kappa_1(t)=t$ e integrando, obtenemos
$$\theta(t)=\int \kappa_1(t)dt=\int tdt$$

```python
reset()
var('t')
kappa1(t) = t

theta(t) = integral(kappa1(t),t)
theta(t).show()
```

Por tanto,
$$\theta(t)=\frac{1}{2}t^2+C$$
Pero sabemos que dada la condición inicial $\textbf{x}_1(a)=(0,0)$ y $\dot{\textbf{x}}_1(a)=(1,0)$, sabemos que $\theta(0)$=0. Por ende, $\theta(0)=1/2\cdot 0^2 + C = 0$. Así que $C=0$ y llegamos a que
$$\theta(t)=\frac{1}{2}t^2$$

Ahora hallamos $x_1(t),y_1(t)$ integrando $\theta(t)$:
$$x_1(t)=\int \cos(\theta(t))dt=\int\cos\left(\frac{1}{2}t^2\right)dt$$
$$x_2(t)=\int \sin(\theta(t))dt=\int\sin\left(\frac{1}{2}t^2\right)dt$$
Démonos cuenta de que estas integrales son justamente las integrales de Fresnel. Las calculamos usando SageMath:

```python
t = var('t')

x1(t) = fresnel_cos(1/2*t**2)
y1(t) = fresnel_sin(1/2*t**2)

p1 = parametric_plot((x1(t), y1(t)), (t, -10, 10), color='blue', thickness=1)

p1.show(figsize=5)
```

Análogamente graficamos $\textbf{x}_2$:

```python
# Para kappa_2 = t + 1
x2(t) = fresnel_cos((t + 0.5*t^2))
y2(t) = fresnel_sin((t + 0.5*t^2))

p2 = parametric_plot((x2(t), y2(t)), (t, -10, 10), color='red', thickness=1)

p2.show(figsize=5)
```
