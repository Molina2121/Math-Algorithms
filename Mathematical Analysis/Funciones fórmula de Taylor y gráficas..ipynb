<p><span style='color:blue;'>Ejercicio 1.</span> Ejecutar la orden taylor? para obtener información sobre esta orden. Comprobar que el polinomio de Taylor de orden 3, en 0, de la función $f(x)=\frac{\sqrt{1+e^x}}{\sqrt{2}\cos x}$, es $1 + \frac{1}{4} x + \frac{19}{32} x^2 + \frac{55}{384} x^3$.</p>

```python
reset()
var('x')
pretty_print_default('True')
f(x)=sqrt(1+e**x)/sqrt(2)/cos(x)
f(x).show()
P(x)=taylor(f(x),x,0,3)
P(x).show()
```

```python
gf=plot(f(x),(x,-1,1),color='red')
gP=plot(P(x),(x,-1,1),color='blue')
```

```python
(gP+gf).show(figsize=6)
```

<p><span style='color:blue;'>Ejercicio 5.</span> Estudiar el dominio, el crecimiento, los extremos, la convexidad y las asíntotas (horizontales u oblicuas) de la función $f(x)=\sqrt{4x^2 - x}$. Representar la función en algún intervalo <em>razonable</em>, junto con sus asíntotas (si las tiene).</p>

```python
reset()
var('x')
f(x)=sqrt(4*x**2-x)
solve(4*x**2-x>0,x)
```

```python
f1(x)= diff(f(x),x)
pc = solve(f1(x)==0,x)
show(pc)
```

```python
A=plot(f(x))
a=lim((f(x)/x),x=oo)
b=lim(f(x)-a*x,x=oo)
ast(x)=a*x+b
B=plot(ast(x),color='red')
a=lim((f(x)/x),x=-oo)
b=lim(f(x)-a*x,x=-oo)
ast2(x)=a*x+b
C=plot(ast2(x),color='green')
```

```python
(A+B+C).show()
```
