<p><span style='color:blue;'>Ejercicio 7:</span> dibujar en una misma figura las gráficas de las cuatro funciones $f(x)=x^2$, $g(x)=2x^2$, $h(x)=\sqrt{x}$ y $j(x)=\sqrt{\frac{x}{2}}$ en algún intervalo a elegir. Hay una región del plano limitada por las cuatro gráficas; rehacer la figura de modo que esta región se vea claramente. Calcular los cuatro puntos de corte entre las funciones que limitan la región.</p>

```python
reset()
var ('x')
f=plot(x**2, (x, -4, 4), color ='yellow')
g=plot(2*x**2, (x, -4, 4), color ='red')
h=plot(sqrt(x), (x, -4, 4), color ='blue')
j=plot(sqrt(x/2), (x, -4, 4), color ='green')
(f+g+h+j).show(figsize=5)
```

<p><span style='color:blue;'>Ejercicio 8:</span> dibujar en una misma figura, con ayuda de la orden implicit_plot, la circunferencia $x^2 + y^2 = 1$, junto con las parábolas $x=y^2$ e $y=2x^2$, en la región $-1 \leq x \leq 1$, $-1 \leq y \leq 1$. Comprobar que hay <span style='color:red;'>cinco</span> regiones limitadas (cada una de ellas) por las tres curvas. Calcular todos los puntos de corte de las curvas y expresarlos con al menos ocho cifras significativas.</p>

```python
reset()
var ('x', 'y')
f=implicit_plot(x**2 + y**2 == 1,(x, -1, 1), (y, -1, 1), color ='yellow')
g=implicit_plot(x == y**2,(x, -1, 1), (y, -1, 1), color ='red')
h=implicit_plot(y == 2*x**2,(x, -1, 1), (y, -1, 1), color ='blue')
(f+g+h).show(figsize=6)
```
